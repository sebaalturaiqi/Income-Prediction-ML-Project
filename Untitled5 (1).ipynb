# Step 1: Environment Setup and Library Imports

```python
!pip install autogluon
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
%matplotlib inline
sns.set(style="whitegrid")
```

# Step 2: Data Loading & Initial Sampling

```python
train_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/train.csv')
test_data = TabularDataset('https://autogluon.s3.amazonaws.com/datasets/Inc/test.csv')

print("SAMPLE OF DATA")
display(train_data.head())

plt.figure(figsize=(6,4))
train_data['class'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Income Class Distribution (<=50K vs >50K)')
plt.show()
```



*   Data Acquisition: The dataset was successfully loaded from the cloud source, containing features like age, workclass, and education.
* Target Variable: The class column is our target, where we aim to predict if income is <=50K or >50K.
*   Income Distribution: * 76% of people are in the low-income bracket, while 24% are in the high-income bracket.
*   Data Balance: * The data is imbalanced, which means the model must be carefully trained to recognize the high-income class.


```python
print(f"Number of rows: {train_data.shape[0]}")
print(f"Number of columns: {train_data.shape[1]}")
print(f"Missing values: {train_data.isnull().sum().sum()}")
```

# Data Quantity and Quality
Requirement: What is the quantity and quality of the data?

The dataset consists of approximately 32,561 records and 14 features. The data quality is high, although it contains some missing values (represented by ?). These missing values were handled automatically by the AutoGluon preprocessing pipeline, ensuring a robust training process.

# Step 3: Model Leaderboard & Performance Ranking

```python
label = 'class'
print("Summary of class variable: \n", train_data[label].describe())
predictor = TabularPredictor(label=label).fit(train_data, time_limit=300)
predictor.leaderboard(silent=True)
```



*   Top Performer: The WeightedEnsemble_L2 achieved the highest accuracy of 88.68% by combining the strengths of multiple models.
*   Efficiency vs. Accuracy: While LightGBM was significantly faster to train (1.85s), the Ensemble model provided the most reliable predictions.



# Model Training

```python
importance = predictor.feature_importance(train_data)
print(importance)
```

# Analysis of Predictive Power
Based on the feature importance results:
1. **Primary Drivers:** `Relationship`, `Education-num`, and `Age` are the most significant predictors for income level.
2. **Socio-Economic Impact:** The model relies on educational attainment and household role more than individual traits like race or nationality, which indicates a logical alignment with economic trends.
3. **Model Transparency:** By identifying these features, we ensure that the model's decisions are explainable and grounded in relevant data.

# Step 4: Feature Importance Analysis

```python
importance = predictor.feature_importance(test_data)
plt.figure(figsize=(10, 6))
importance['importance'][:10].plot(kind='barh', color='teal')
plt.title('Top 10 Features Driving Income Prediction')
plt.xlabel('Importance Score')
plt.ylabel('Feature Name')
plt.gca().invert_yaxis()
plt.show()
```



*   Top Predictor: Capital-gain is the most significant feature, meaning the model relies heavily on investment profits to identify high-income individuals.
*   Key Factors: Marital status and Age are the next most important features, showing that social stability and work experience are strong indicators of income levels.



# Step 5: Final Model Evaluation (Confusion Matrix)

```python
performance = predictor.evaluate(test_data)
print("FINAL PERFORMANCE RESULTS:")
print(performance)

y_pred = predictor.predict(test_data)
y_true = test_data[label]
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=predictor.class_labels)

fig, ax = plt.subplots(figsize=(8, 6))
disp.plot(cmap='Blues', ax=ax)
plt.title('Confusion Matrix: Performance Assessment')
plt.show()
```



*  Accuracy Breakdown: The model correctly predicted 7,040 low-income cases and 1,505 high-income cases on new, unseen data.
* Robust Performance: Overall accuracy on the test set is approximately 87.5%, confirming the model's readiness for real-world use.



# Step 6: Interactive Income Predictor Application

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd


age_input = widgets.IntSlider(value=30, min=17, max=90, description='Age:')
work_input = widgets.Dropdown(options=[' Private', ' Self-emp-not-inc', ' Local-gov', ' State-gov'], value=' Private', description='Workclass:')
edu_input = widgets.Dropdown(options=[' Bachelors', ' HS-grad', ' Masters', ' Doctorate', ' Some-college'], value=' Bachelors', description='Education:')
cap_gain_input = widgets.IntText(value=0, description='Cap. Gain:')
hours_input = widgets.IntSlider(value=40, min=1, max=99, description='Hours/Week:')

predict_button = widgets.Button(description='Predict Income', button_style='success')
output = widgets.Output()


def on_predict_clicked(b):
    with output:
        clear_output()

        input_data = pd.DataFrame([{
            'age': age_input.value,
            'workclass': work_input.value,
            'fnlwgt': 200000,
            'education': edu_input.value,
            'education-num': 13 if edu_input.value == ' Bachelors' else 9,
            'marital-status': ' Married-civ-spouse',
            'occupation': ' Exec-managerial',
            'relationship': ' Husband',
            'race': ' White',
            'sex': ' Male',
            'capital-gain': cap_gain_input.value,
            'capital-loss': 0,
            'hours-per-week': hours_input.value,
            'native-country': ' United-States'
        }])


        res = predictor.predict(input_data)[0]
        prob = predictor.predict_proba(input_data).iloc[0].max()


        color = "green" if res == ' >50K' else "red"
        print(f"--- Result ---")
        display(widgets.HTML(f"<h2 style='color:{color}'>Prediction: {res}</h2>"))
        print(f"Confidence Level: {prob:.2%}")

predict_button.on_click(on_predict_clicked)


print("--- Simple Income Predictor App ---")
ui = widgets.VBox([age_input, work_input, edu_input, cap_gain_input, hours_input, predict_button, output])
display(ui)
```



*   Real-Time Prediction: Created a user-friendly interface that allows users to input personal data and get instant predictions from the trained model.
